```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://babybathwater.com/members/'
```

```python
page = requests.get(url)
```

```python
page
```

```python
html = page.content
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
members = soup.find_all('div', class_='member-title-box')
members
```

```python
master_list = []
for member in members:
    member_dict = {}
    member_dict['Name'] = member.find('div', class_='title')
    member_dict['website'] = member.find('a', href= True)
    
    master_list.append(master_list)
    
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(master_list)
```
